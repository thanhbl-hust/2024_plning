```python
def findMaxandMin(ls):
    maxIncrease = -10
    inroot1 = -1
    inroot2 = -1
    maxDecrease = -10
    deroot1 = -1
    deroot2 = -1
    for i in range(1, len(ls)):
        if ls[i] > ls[i - 1]:
            if ls[i] - ls[i - 1] > maxIncrease:
                maxIncrease = ls[i] - ls[i - 1]
                inroot1 = i - 1
                inroot2 = i 
        elif ls[i] < ls[i - 1]:
            if ls[i - 1] - ls[i] > maxDecrease:
                maxDecrease = ls[i - 1] - ls[i]
                deroot1 = i - 1
                deroot2 = i
    return maxIncrease, maxDecrease, inroot1, inroot2, deroot1, deroot2
```

```python
def paint(w, d, set_time):


    times = set_time
    times_appear = len(d) / times
    times_appear = int(times_appear)
    n = len(w)
    average_values = []
    for i in range(0, n, times_appear):
        average = np.mean(w[i: i + times_appear])
        average_values.append(average)
    x_ticks = range(0, n, times_appear)
    x_labels = [d[i] for i in x_ticks]

    max_value = np.max(w)
    min_value = np.min(w)

    #oneday has 2 parameters, 0:00 and after wake up
    maxIn, maxDe, inr1, inr2, der1, der2 = findMaxandMin(w)

    print("INF OF PROCESS:")
    print("\n")
    print(f'Day of process: {int((len(w) + len(w) % 2) / 2)} days')
    print(f'From: {d[0]} to now: {d[len(d) - 1]}')
    print(f'Max weight in process: {round(max(w), 2)} (kg)')
    print(f'Min weight in process: {round(min(w), 2)} (kg)')
    print(f'Max increase in one process: {round(maxIn, 1)}___date:{d[inr1]} - {d[inr2]} ({w[inr1]} to {w[inr2]})')
    print(f'Max decrease in one process: {round(maxDe, 2)}___date:{d[der1]} - {d[der2]} ({w[der1]} to {w[der2]})')
    print(f'Total weight decrease: {round(abs(w[len(w) - 1] - w[0]), 1)} kg after {len(w) - 1} process ({w[0]} - {w[len(w) - 1]}).')

    fig = plt.figure(figsize=(15, 10))
    plt.axhline(y = max_value, color = 'r', linestyle = '--', label = 'Max')
    plt.axhline(y = min_value, color = 'g', linestyle = '--', label = 'Min')
    plt.plot(d, w, marker = '.')
    plt.xlabel('Time (D: day, N: night)')
    plt.ylabel('W (kg)')
    plt.title('W by Ds')
    plt.xticks(x_ticks, x_labels, rotation = 45)
    plt.legend(['max weight', 'min weight'], loc='lower left')
    plt.show()


check = len(w)
paint(w[:check], d[:check], 11)
```
